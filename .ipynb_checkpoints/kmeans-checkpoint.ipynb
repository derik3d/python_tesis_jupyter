```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler



X_std = StandardScaler().fit_transform(train_df_desc)

# Create a PCA instance: pca
pca = PCA()

principalComponents = pca.fit_transform(X_std)
# Plot the explained variances

features = range(pca.n_components_)





plt.bar(features, pca.explained_variance_ratio_, color='black')
plt.xlabel('PCA features')
plt.ylabel('variance %')
plt.xticks(features)
# Save components to a DataFrame
PCA_components = pd.DataFrame(principalComponents)
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

```python
plt.scatter(PCA_components[0], PCA_components[1], alpha=.1, color='black')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(PCA_components[0], PCA_components[1], PCA_components[2],c=train_df_group.iloc[:,9].values)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
ks = range(2, 60)
inertias = []
for k in ks:
    # Create a KMeans instance with k clusters: model
    model = KMeans(n_clusters=k)
    
    # Fit model to samples
    model.fit(PCA_components)
    
    
    # Append the inertia to the list of inertias
    inertias.append(model.inertia_)
    
plt.plot(ks, inertias, '-o', color='black')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

```python
from sklearn.metrics import  silhouette_score
silhouette_scores = []
K = range(2,60)
k_clus = []
X=PCA_components

for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(X)
    y = km.predict(X)
    silhouette_scores.append(silhouette_score(X, y))
    k_clus.append(k)
    
plt.plot(k_clus, silhouette_scores, '-o', color='black')
plt.xlabel('k')
plt.ylabel('sil')
plt.xticks(ks)
plt.show()
```

```python
#for n_clusters in range_n_clusters:
sil = []
k_clus = []

for k in ks:
    clusterer = KMeans(n_clusters=k)
    preds = clusterer.fit_predict(PCA_components)
    centers = clusterer.cluster_centers_

    score = silhouette_score(PCA_components, preds)
    #print("For n_clusters = {}, silhouette score is {})".format(k, score))
    k_clus.append(k)
    sil.append(score)
    
    
    
plt.plot(k_clus, sil, '-o', color='black')
plt.xlabel('k')
plt.ylabel('sil')
plt.xticks(ks)
plt.show()
```

```python
import numpy as np
from scipy.spatial.distance import pdist
from scipy.spatial.distance import directed_hausdorff


def dfun(u, v):
    
    u_n_n_list = df_node_names.loc[df_node_names['cpd_id'].isin(df_cpd_rct[df_cpd_rct.rct_id.eq(u[0])]["cpd_id"])]['cpd_name']
    #print(u_n_n_list)
    #print(df_desc[df_desc.index.isin(u_n_n_list)])
    #print(df_desc.loc[u_n_n_list])
    v_n_n_list = df_node_names.loc[df_node_names['cpd_id'].isin(df_cpd_rct[df_cpd_rct.rct_id.eq(v[0])]["cpd_id"])]['cpd_name']
    #print(v_n_n_list)
    #print(df_desc[df_desc.index.isin(v_n_n_list)])
    #print(df_desc.loc[v_n_n_list])
    #print("")
    try:
        ans =  directed_hausdorff(df_desc.loc[u_n_n_list], df_desc.loc[v_n_n_list])[0]
    except:
        ans = 10
    
    
    return ans

dm = pdist(df_cpd_rct_enabled_nodes.iloc[:50,:], dfun)

dm

#dfun(np.array([2513]),np.array([2511]))
```
